```python
import sqlite3
vitamin_names = []


def search_records(vitamin_names):

    conn = sqlite3.connect('vitamin.db')
    cursor = conn.cursor()


    conditions = " AND ".join([f"{vitamin_name} > 0" for vitamin_name in vitamin_names])
    query = f"SELECT category FROM table1 WHERE {conditions}"


    cursor.execute(query)


    rows = cursor.fetchall()
    for row in rows:
        print(row)

    conn.close()

n=10
for i in range(n):
    vitamin_name = input("enter vitamin name=")
    if(vitamin_name == "done"):
       break
    vitamin_names.append(vitamin_name)




search_records(vitamin_names)
```
